En primera instancia, vamos a definir los datos de prueba que se van a utilizar

```python
import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
#Se grafican los datos para validar la relacion

plt.scatter(x,y)
plt.show()

#Se visualiza que los datos tienen una relacion lineal negativa.
#Es decir, que mientras los valores en x aumentan, los valores de y van disminuyendo
```

Posteriormente, al saber que la relacion entre las variables es lineal, podemos utilizar el algoritmo de regresion lineal simple, ya que son una variable dependiente(y)

y una variable independiente(x)

```python
#Primero se importa el modulo stats de la libreria scipy

from scipy import stats

#En esta libreria se encuentra la funcion que se necesitará para obtener los coeficientes necesarios para aplicar la formula de regresion lineal
#Esta formula es:   y = intersepcion + pendiente * x

pendiente, intercepcion, r, p, std_err = stats.linregress(x,y)

#La funcion stats.linregress() devuelve los siguientes valores
#pendiente = es la pendiente que se utilizara para dibujar la linea de regresion (aumento en y por cada unidad de aumento en x)
#intercepcion = es la intercepcion que se utilizara para dibujar la linea de regresion (valor de y cuando x es igual a 0)
#r = coeficiente de correlacion. Esto indica que tan relacionadas estan las variables toma valores de -1 a 1. donde ambos extremos indican una fuerte relacion.
#p = valor p. Evalua la significancia estadistica de la relacion
```

```python
#Posteriormente al tener estos valores verificamos:
    #1. vemos el valor de r para verificar la relacion entre ambas variables
    #2. Vemos si esta relacion es estadisticamente significativa o es producto del azar

print("El valor de correlacion es", r)

print('El valor de significancia estadistica es', p)
```

Si el valor de correlacion esta lejos de 0 y se acerca a cualquiera de los dos extremos(-1 y 1) quiere decir que tiene una correlacion bastante fuerta

Si el valor p o significancia estadistica es menor a 0.05 indica que tiene mucha significancia y solidez la relacion.

Ahora sabiendo esto, podemos utilizar los otros dos valores para calcular la linea de regresion.

```python
#Vamos a realizar una funcion con la formula de la regresion, y pasaremos cada valor de x por esta funcion.

def regresion_lineal(x):
    return intercepcion + pendiente * x

#Ahora utilizaremos la funcion map() para pasar cada valor de x por la funcion que se acaba de crear, y esos valores se transformaran en una lista que se llamara "modelo"

modelo = list(map(regresion_lineal, x))
```

Ahora que se tiene el modelo entrenado, y con sus respectivas predicciones, vamos a visualizarlo en el grafico

```python
plt.scatter(x,y)
plt.plot(x, modelo)
plt.title('regresion lineal')
plt.show()
```

### Ahora con este modelo entrenado, vamos a utilizar nuevos valores en x para verificar que tan bien realiza las predicciones.

Esto se realizara reemplazando ciertos valores en la lista actual.
